```python
# Checking the missing values in the dataframe
df.isnull().sum()
```

```python
# Replacing the missing values by category-wise median data
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# Checking for the missing values in the dataframe
df.isnull().sum()
```

```python
# Changing the column names into lower case
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df = df.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns
```

```python
# Creating age_group column
lables = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q = 4, labels = lables)
```

```python
df[['age', 'age_group']].head(10)
```

```python
# Creating purchase_frequency_days column
frequency_days = {
    'Annually' : 365,
    'Bi-Weekly' : 14,
    'Every 3 Months' : 90,
    'Fortnightly' : 14,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Weekly' : 7
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_days)
```
